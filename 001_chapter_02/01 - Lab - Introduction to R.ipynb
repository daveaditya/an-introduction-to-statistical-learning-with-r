## 2.3.2 Graphics

```r
?plot
```

```r
x = rnorm(100)
y = rnorm(100)
plot(x, y)
```

```r
plot(x, y, xlab="this is the x axis", ylab="this is the y axis", main="Plot of X vs Y")
```

Use the `pdf()` function to save as PDF. And `jpeg()` function to save as JPEG.

```r
pdf("figure.pdf")
plot(x, y, col="green")
dev.off()
```

The `dev.off()` function indicates that we are done plotting the graph

The `seq()` function can be used to create a sequence of numbers. For integer arguments `:` operator can be used as a shorthand for example: 1:10 stands for seq(1, 10)

```r
x = seq(1, 10)
x
```

```r
1:10
```

```r
pi
```

```r
x = seq(-pi, pi, length=50)
x
```

The `contour()` function produces a contour plot in order to represent three-dimensional data; it is like a topographical map

```r
?contour
```

```r
y=x
f = outer (x ,y , function (x , y ) cos ( y ) /(1+ x ^2) )
contour (x ,y , f )
contour (x ,y ,f , nlevels =45 , add = T )
fa =( f - t ( f ) ) /2
contour (x ,y , fa , nlevels =15)
```

The `image()` function works the same way as `contour()` except that it produces a color-coded plot whose colors depend on the z value

```r
image(x, y, fa)
```

`persp()` can be used to produce a three-dimensional plot

```r
persp(x, y, fa)
```

```r
persp(x, y, fa, theta=30)
```

```r
persp(x, y, fa, theta=30, phi=20)
```

```r
persp(x, y, fa, theta=30, phi=70)
```

```r
persp(x, y, fa, theta=30, phi=40)
```

## 2.3.3 Indexing Data

```r
A = matrix(1:16, nrow=4, ncol=4)
A
```

```r
A[2, 3]
```

 The first number after the open-bracket symbol `[` always refers to the row, and the second number always refers to the column.

```r
A[c(1, 3), c(2, 4)]
```

```r
A[1:3, 2:4]
```

```r
A[1:2,]
```

```r
A[,1:2]
```

```r
A[1,]
```

The use of a negative sign - in the index tells R to keep all rows or columns except those indicated in the index.

```r
A[-c(1,3),]
```

The `dim()` function outputs the number of rows followed by the number of columns of a given matrix

```r
dim(A)
```

## 2.3.5 Additional Graphical and Numerical Summaries 

To refer to a variable, we must type the data set and the variable name
joined with a `$` symbol

```r
plot(Auto$cylinders, Auto$mpg)
```

We can use the `attach()` function in order to tell in `R`  to make the variables in a data frame available by name

```r
attach(Auto)
```

```r
plot(cylinders, mpg)
```

The `as.factor()` function converts quantitative variables into qualitative variables

```r
cylinders = as.factor(cylinders)
```

If the variable plotted on x-axis is categorical, the `plot()` function automatically creates `boxplots`

```r
plot(cylinders, mpg)
```

```r
plot(cylinders, mpg, col="red")
```

```r
plot(cylinders, mpg, col="red", varwidth=TRUE)
```

```r
plot(cylinders, mpg, col="red", varwidth=TRUE, horizontal=TRUE)
```

```r
plot(cylinders, mpg, col="red", varwidth=TRUE, horizontal=TRUE, xlab="cylinders", ylab="MPG")
```

The `hist()` function is used to plot histogram

```r
hist(mpg)
```

```r
hist(mpg, col=2)
```

```r
hist(mpg, col=2, breaks=15)
```

The `pairs()` function creates a scatterplot matrix, i.e. a scatterplot for every pair of variables for any given data set

```r
pairs(Auto)
```

```r
pairs (~ mpg + displacement + horsepower + weight + acceleration , Auto)
```

In conjunction with the `plot()` function, `identify()` provides a useful interactive method for identifying the value for a particular variable for points on a plot

```r
plot(horsepower, mpg)
identify(horsepower, mpg, name)
```

The `summary()` function produces a numerical summary of each variable in a particular data set

```r
summary(Auto)
```

```r
summary(mpg)
```

Once we have finished using R, we type `q()` in order to shut it down, or
quit. When exiting R, we have the option to save the current workspace so
workspacethat all objects (such as data sets) that we have created in this R session
will be available next time. Before exiting R, we may want to save a record
of all of the commands that we typed in the most recent session; this can
be accomplished using the `savehistory()` function. Next time we enter R,
we can load that history using the `loadhistory()` function.
